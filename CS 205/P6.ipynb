*Create a graph to show the popularity of a few names over the years*

```python
import pandas as pd
from plotnine import *
BN = pd.read_csv("https://mcsp.wartburg.edu/letsche/cs205/BabyNames-2022.csv")
BN = BN[BN['name'].isin(['Andrew', 'Eli', 'Paul'])]
(ggplot(BN, aes(x='year', y='count', group='name', color='name')) +
geom_line(size=1, alpha=0.5) +
labs(y="Popularity", x="Year"))
```
